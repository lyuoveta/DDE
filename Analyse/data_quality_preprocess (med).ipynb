```python
import pandas as pd
data = pd.read_csv(r'D:\Downloads\combined_data (1).csv')
cleaned_data = pd.read_csv(r'D:\Downloads\cleaned_data.csv')
```

```python
data['dates'] = pd.to_datetime(data['dates'])
data.loc[:,'age'] = pd.to_numeric(data['age'], errors='coerce')
data.loc[:,'duration, months'] = pd.to_numeric(data['duration, months'], errors='coerce')
```

```python
data.info()
```

```python
from ydata_profiling import ProfileReport
# from .autonotebook import tqdm as notebook_tqdm

profile = ProfileReport(data, title="Profiling Report")
```

```python
import ipywidgets

profile.to_file("original.html")
```

```python
cleaned_data['dates'] = pd.to_datetime(cleaned_data['dates'])
```

```python
cleaned_data.info()
```

```python
profile_clean = ProfileReport(cleaned_data, title="Profiling Report")
```

```python
comparison_report = profile.compare(profile_clean)
comparison_report.to_file("original_vs_transformed.html")
```
